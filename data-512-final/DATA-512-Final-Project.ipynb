## Background/Related Work
The concept of food desert has been around for more than a decade and its definition was documented here in the report [Starved for access: life in rural America's food deserts](https://www.ruralsociology.org/assets/docs/rural-realities/rural-realities-1-4.pdf) from 2007. A food desert is an area where people have limited access to a variety of healthy and affordable food. More formally, the [United States Department of Agriculture (USDA)](https://www.ers.usda.gov/data-products/food-access-research-atlas/documentation/) defines a food desert as a Census Tract with substantial low income populations that have low access to affordable, healthful food retail such as supermarkets. Quantitatively, the datasets defined regions as foods deserts using three different measuring standards:

1.	Low-income and low-access tract measured at 1 mile and 10 miles Definition: A low-income tract with at least 500 people, or 33 percent of the population, living more than 1 mile (urban areas) or more than 10 miles (rural areas) from the nearest supermarket, supercenter, or large grocery store. Using this measure, an estimated 54.4 million people, or 17.7 percent of the U.S. population, live in tracts that are low-income and low access and are more than ½ mile or 10 miles from the nearest supermarket.
2.	Low-income and low-access tract measured at ½ mile and 10 miles Definition: A low-income tract with at least 500 people, or 33 percent of the population, living more than ½ mile (urban areas) or more than 10 miles (rural areas) from the nearest supermarket, supercenter, or large grocery store. This measure shows that an estimated 19 million people, or 6.2 percent of the U.S. population, live in low-income and low access tracts and are more than 1 mile or 10 miles from a supermarket.
3.	Low-income and low-access tract measured at 1 mile and 20 miles Definition: A low-income tract with at least 500 people, or 33 percent of the population, living more than 1 mile (urban areas) or more than 20 miles (rural areas) from the nearest supermarket, supercenter, or large grocery store. Under this measure, an estimated 17.3 million people, or 5.6 percent of the U.S. population, live in low-income and low access tracts and are more than 1 mile or 20 miles from a supermarket.

There has been a lot of effort into understanding where food deserts are and USDA has published their census data and created an [Food Access Research Atlas](https://www.ers.usda.gov/data-products/food-access-research-atlas/go-to-the-atlas/) where people can interactively map the data and visualize them.

Throughout the years, researchers have linked the lack of healthy and affordable food options with public health problems, especially chronic diet-related diseases such as diabetes and obesity. In 2009, the National Academy of Science published a detailed [report](https://www.nap.edu/catalog/12623/the-public-health-effects-of-food-deserts-workshop-summary) with an overview of the extend of food deserts, studies of food deserts from multiple aspects, as well as evidence on how limited food access can affect people's diet and health. The report showed that the problem of food deserts was relevant, but it was from more than a decade ago. I would like to work with the latest data from the Food Access Research Atlas to see if the problems persist.

```python
# import necessary packages
import pandas as pd
from scipy.stats import ttest_ind
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats.stats import pearsonr
import geopandas as gpd
import geoplot as gplt
import geoplot.crs as gcrs
import mapclassify as mc
```

```python
# read in raw csv files as pandas dataframes
food_access = pd.read_csv("./data/raw/food_access.csv")
physical_health = pd.read_csv("./data/raw/Nutrition__Physical_Activity__and_Obesity_-_Behavioral_Risk_Factor_Surveillance_System.csv")
```

### Data Preparation 

```python
# select useful columns to group at the county level
df_low_food_access_pop = pd.DataFrame(food_access.groupby(['State', 'County', 'Urban'])
                                      ['lapop1', 'lapop10', 'LAPOP1_10', 'POP2010', 
                                       'lalowi1', 'lalowi10',
                                       'lakids1', 'lakids10', 'laseniors1', 'laseniors10',
                                       'lawhite1', 'lawhite10', 'lablack1', 'lablack10', 
                                       'laasian1', 'laasian10', 'lahisp1', 'lahisp10',
                                       'TractLOWI', 'TractKids', 'TractSeniors', 
                                       'TractWhite', 'TractBlack', 
                                       'TractAsian', 'TractHispanic'].sum()).reset_index()

df_income = pd.DataFrame(food_access.groupby(['State', 'County', 'Urban'])[
    'PovertyRate','MedianFamilyIncome'].mean()).reset_index()
```

```python
# calculate ratios for low food access population in general and for each race
df_low_food_access_pop['la1_ratio'] = df_low_food_access_pop['lapop1'] / df_low_food_access_pop['POP2010']
df_low_food_access_pop['la10_ratio'] = df_low_food_access_pop['lapop10'] / df_low_food_access_pop['POP2010']
df_low_food_access_pop['la_ratio'] = df_low_food_access_pop['LAPOP1_10']/ df_low_food_access_pop['POP2010']
df_low_food_access_pop['la_percentage'] = df_low_food_access_pop['la_ratio']*100
df_low_food_access_pop['pop_ratio_poverty'] = df_low_food_access_pop['TractLOWI'] / df_low_food_access_pop['POP2010']
df_low_food_access_pop['pop_ratio_kid'] = df_low_food_access_pop['TractKids'] / df_low_food_access_pop['POP2010']
df_low_food_access_pop['pop_ratio_senior'] = df_low_food_access_pop['TractSeniors'] / df_low_food_access_pop['POP2010']
df_low_food_access_pop['pop_ratio_white'] = df_low_food_access_pop['TractWhite'] / df_low_food_access_pop['POP2010']
df_low_food_access_pop['pop_ratio_black'] = df_low_food_access_pop['TractBlack'] / df_low_food_access_pop['POP2010']
df_low_food_access_pop['pop_ratio_asian'] = df_low_food_access_pop['TractAsian'] / df_low_food_access_pop['POP2010']
df_low_food_access_pop['pop_ratio_hisp'] = df_low_food_access_pop['TractHispanic'] / df_low_food_access_pop['POP2010']

df_low_food_access_pop['la_ratio_poverty'] = (df_low_food_access_pop['lalowi1'] +  df_low_food_access_pop['lalowi10'])/(df_low_food_access_pop['lapop1'] + df_low_food_access_pop['lapop10'])
df_low_food_access_pop['la_ratio_kid'] = (df_low_food_access_pop['lakids1'] +  df_low_food_access_pop['lakids10']) / (df_low_food_access_pop['lapop1'] + df_low_food_access_pop['lapop10'])
df_low_food_access_pop['la_ratio_senior'] = (df_low_food_access_pop['laseniors1'] +  df_low_food_access_pop['laseniors10']) / (df_low_food_access_pop['lapop1'] + df_low_food_access_pop['lapop10'])
df_low_food_access_pop['la_ratio_white'] = (df_low_food_access_pop['lawhite1'] +  df_low_food_access_pop['lawhite10']) / (df_low_food_access_pop['lapop1'] + df_low_food_access_pop['lapop10'])
df_low_food_access_pop['la_ratio_black'] = (df_low_food_access_pop['lablack1'] +  df_low_food_access_pop['lablack10']) / (df_low_food_access_pop['lapop1'] + df_low_food_access_pop['lapop10'])
df_low_food_access_pop['la_ratio_asian'] = (df_low_food_access_pop['laasian1'] +  df_low_food_access_pop['laasian10']) / (df_low_food_access_pop['lapop1'] + df_low_food_access_pop['lapop10'])
df_low_food_access_pop['la_ratio_hisp'] = (df_low_food_access_pop['lahisp1'] +  df_low_food_access_pop['lahisp10']) / (df_low_food_access_pop['lapop1'] + df_low_food_access_pop['lapop10'])
```

```python
# filter out wrong data where the number of low food access population exceeds the total population
df_low_food_access_pop = df_low_food_access_pop[(df_low_food_access_pop['la_ratio']<1)&
                                                (df_low_food_access_pop['pop_ratio_poverty']<1)]
```

```python
df_la = df_low_food_access_pop.merge(df_income, left_on = ['State', 'County', 'Urban'], right_on = ['State', 'County', 'Urban'])
```

```python
# low food access data summary
df_la[['la_ratio','pop_ratio_white', 'pop_ratio_black', 'pop_ratio_asian', 'pop_ratio_hisp']].describe()
```

```python
# select time range and questions to include for the physical health data
df_health_2010 = physical_health[physical_health['YearStart'] == 2011]
df_health_final = df_health_2010[((df_health_2010['Question'] == 'Percent of adults aged 18 years and older who have obesity')|
                                 (df_health_2010['Question'] == 'Percent of adults aged 18 years and older who have an overweight classification')|
                                 (df_health_2010['Question'] == 'Percent of adults who engage in no leisure-time physical activity'))&
               (df_health_2010['Stratification1'] == 'Total')&(df_health_2010['LocationAbbr'] != 'US')][[
    'GeoLocation', 'LocationAbbr', 'LocationDesc', 'Question','Data_Value']].reset_index()
```

```python
# merge low food access data with physical health data
df_food_by_state = pd.DataFrame(df_low_food_access_pop.groupby(['State'])['la_ratio'].mean()).reset_index()
df_merge = df_health_final.merge(df_food_by_state, left_on = 'LocationDesc', right_on = 'State')
```

```python
# save final dataframes as csv files
df_merge.to_csv('./data/final/food_access_health_merge.csv')
df_la.to_csv('./data/final/food_access.csv')
```

### Question 3: How does food access limit affect physical health measured by obesity rate?

```python
contiguous_usa['Overweight_percent'] = contiguous_usa['state'].map(
    lambda state: df_la.query("State == @state").iloc[0]['PovertyRate']
)
scheme = mc.Quantiles(contiguous_usa['Overweight_percent'], k=5)

gplt.choropleth(
    contiguous_usa, hue='Overweight_percent', 
    projection=gcrs.AlbersEqualArea(central_longitude=-98, central_latitude=39.5),
    edgecolor='white', linewidth=1,
    cmap='Reds',
    legend=True, legend_kwargs={'loc': 'lower left'},
    scheme=scheme,
    figsize = (12,16)
    
)
plt.title("Overweight, 2010")
```

```python
plt.figure(figsize=(9, 6))
sns.regplot(data = df_merge[df_merge['Question'] == 'Percent of adults aged 18 years and older who have obesity'], 
            x = 'la_ratio', y = 'Data_Value', ci = None)
sns.despine()
x = df_merge[df_merge['Question'] == 'Percent of adults aged 18 years and older who have obesity']['la_ratio']
y = df_merge[df_merge['Question'] == 'Percent of adults aged 18 years and older who have obesity']['Data_Value']
plt.text(0.05,34,"R squared = " + str(np.round(pearsonr(x,y)[0]**2,4)), fontsize = 12)
plt.text(0.05,33.5,"P value = " + str(np.round(pearsonr(x,y)[1],4)), fontsize = 12)


plt.ylabel('Obesity Percentage')
plt.xlabel('Low Food Access Ratio')
plt.title('Obesity vs Food Desert')
```

```python
plt.figure(figsize=(9, 6))
sns.regplot(data = df_merge[df_merge['Question'] == 'Percent of adults aged 18 years and older who have an overweight classification'], 
            x = 'la_ratio', y = 'Data_Value', ci = None)
sns.despine()
x = df_merge[df_merge['Question'] == 'Percent of adults aged 18 years and older who have an overweight classification']['la_ratio']
y = df_merge[df_merge['Question'] == 'Percent of adults aged 18 years and older who have an overweight classification']['Data_Value']
plt.text(0.05,38,"R squared = " + str(np.round((pearsonr(x,y)[0])**2,4)), fontsize = 12)
plt.text(0.05,37.5,"P value = " + str(np.round(pearsonr(x,y)[1],4)), fontsize = 12)

plt.ylabel('Overweight Percentage')
plt.xlabel('Low Food Access Ratio')
plt.title('Overweight vs Food Desert')
```

```python
plt.figure(figsize=(9, 6))
sns.regplot(data = df_merge[df_merge['Question'] == 'Percent of adults who engage in no leisure-time physical activity'], x = 'la_ratio', y = 'Data_Value', ci = None)
sns.despine()
x = df_merge[df_merge['Question'] == 'Percent of adults who engage in no leisure-time physical activity']['la_ratio']
y = df_merge[df_merge['Question'] == 'Percent of adults who engage in no leisure-time physical activity']['Data_Value']
plt.text(0.05,35,"R squared = " + str(np.round((pearsonr(x,y)[0])**2,4)), fontsize = 12)
plt.text(0.05,34,"P value = " + str(np.round(pearsonr(x,y)[1],4)), fontsize = 12)
plt.xlabel('No Activity Percentage')
plt.xlabel('Low Food Access Ratio')
plt.title('No Activity vs Food Desert')
```
